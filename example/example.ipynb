## **2. Nhập các yêu cầu đầu vào**

Các yêu cầu cần nhập bắt buộc để chạy chương trình bao gồm:

- **keyword**: Từ khóa bài viết
- **language**: Ngôn ngữ bài viết
- **article_tine**: Giọng điệu bài viết
- **model**: Mô hình sử dụng (Chỉ hỗ trọ các mô hình của OpenAI)
- **word_count**: Số lượng từ cần viết

````python
from writer_agent.graphs.seo_content_writer import AutomatedSEOContentWriterGraph
from writer_agent.schemas.config import UserConfig
from langchain_openai import ChatOpenAI
from datetime import datetime

KEYWORD = "What is GPT-o1?" # Từ khóa muốn viết bài

user_input = UserConfig(
    user_id="02", # Optional
    agent_name="SEOAutomatedWriter", # Optional
    keyword=KEYWORD,
    language="English", # Ngôn ngữ bài viết
    article_tone="Professional", # Giọng điệu bài viết
    model="gpt-4o-mini", # Mô hình sử dụng (Chỉ hỗ trọ các mô hình của OpenAI)
    word_count=2000, # Số lượng từ cần viết
    additional_prompt=None, # Chưa hỗ trợ
)

graph = AutomatedSEOContentWriterGraph(
    user_config=user_input,
    config={"configurable": {"thread_id": "my-thread"}}
)
````

## **3. Trực quan hóa luồng hoạt động hệ thống**

````python
graph.visulize_graph()
````

## **4. Chạy hệ thống**
